# Gold Layer: Scoperta dei Cluster con K-Means

Questo notebook rappresenta la fase finale e più importante del nostro progetto: il **Gold Layer**. L'obiettivo è passare da una descrizione dettagliata dei dati (il Silver Layer) a una scoperta di pattern tra stazioni

Partendo dalla tabella `profili_stazioni_silver`, che contiene la misurazioni di varie features per ogni stazione di monitoraggio, useremo un algoritmo di clustering non supervisionato (K-Means) per rispondere alla nostra domanda di ricerca principale:

> **Esistono "famiglie" o categorie naturali di stazioni con profili comportamentali statisticamente simili? Se sì, cosa le caratterizza?**

Il processo si articolerà in tre macro-fasi:
1.  **Preparazione dei Dati:** Selezione e normalizzazione delle feature per prepararle all'analisi.
2.  **Ricerca del K Ottimale:** Esecuzione di un esperimento per determinare il numero di cluster più appropriato per i nostri dati.
3.  **Addestramento e Interpretazione:** Creazione del modello finale, assegnazione delle etichette di cluster e analisi qualitativa dei gruppi scoperti.

---
## Fase 1: Setup dell'Ambiente e Caricamento dei Dati

Come primo passo, importiamo tutte le librerie necessarie per l'analisi, il machine learning e la visualizzazione. Configuriamo poi la connessione a MinIO S3 per caricare la tabella `profili_stazioni_silver`, che costituisce il punto di partenza per tutta la nostra analisi. Per sicurezza ho impostato come variabile di ambiente le credenziali per accedere a S3, l'indirizzo ip del server e anche un variabile `IS_REMOTE` che nel caso sia settata a false, dice al notebook di cercare le tabelle in locale

```python
import pandas as pd
import numpy as np
from deltalake import DeltaTable
from deltalake.writer import write_deltalake
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import silhouette_score

import mlflow

import os
from dotenv import load_dotenv

load_dotenv()

print("Inizio Fase 1: Setup e Caricamento Dati Silver...")

# Credenziali e configurazione per accedere a MinIO S3
storage_options = { 
    "AWS_ACCESS_KEY_ID": os.getenv("AWS_ACCESS_KEY_ID"),                
    "AWS_SECRET_ACCESS_KEY": os.getenv("AWS_SECRET_ACCESS_KEY"), 
    "AWS_ENDPOINT_URL": os.getenv("AWS_ENDPOINT_URL"), 
    "AWS_ALLOW_HTTP": "true" 
}

IS_REMOTE = os.getenv("IS_REMOTE", 'false').lower() == 'true'
TABLE_SILVER = "profili_stazioni_silver"

REMOTE_BASE_PATH = "s3a://external/"
LOCAL_BASE_PATH = "../delta_tables/"
DELTA_GOLD_TABLE = os.path.join(LOCAL_BASE_PATH, 'stazioni_selezionate_gold')

df_profiles = None

# Carichiamo la tabella Silver che abbiamo creato nello script precedente
try:

    if IS_REMOTE:
        print(f"Modalità Remota: Caricamento dati da S3 ({REMOTE_BASE_PATH})...")
        df_profiles = DeltaTable(f"{REMOTE_BASE_PATH}{TABLE_SILVER}", storage_options=storage_options).to_pandas()
    else:
        print(f"Modalità Locale: Caricamento dati da cartella locale ({LOCAL_BASE_PATH})...")
        df_profiles = DeltaTable(f"{LOCAL_BASE_PATH}{TABLE_SILVER}", storage_options=storage_options).to_pandas()

    
    df_profiles.set_index('id_stazione', inplace=True)
    print(f"Tabella Silver caricata con successo. Contiene {len(df_profiles)} profili di stazioni.")
    display(df_profiles.head())
except Exception as e:
    print(f"ERRORE CRITICO: Impossibile caricare la tabella Silver. Dettagli: {e}")
```

---
## Fase 2: Preparazione dei Dati per il Machine Learning

Un algoritmo come K-Means non può lavorare direttamente sui dati grezzi della tabella Silver. Dobbiamo eseguire due passaggi di pre-processing fondamentali per garantirne il corretto funzionamento:

1.  **Selezione delle Feature:** Isoleremo solo le colonne numeriche che rappresentano le caratteristiche comportamentali delle stazioni (es. `media_PM10`, `rapporto_feriale_festivo_NO2`, etc.). Le colonne contenenti metadati descrittivi (come `Stazione`, `COMUNE`) verranno escluse da questa fase, poiché l'algoritmo non può interpretarle.

2.  **Normalizzazione:** Questo è il passaggio tecnico più critico. Le nostre feature si trovano su scale molto diverse (una media di PM10 può valere 40, mentre una correlazione è compresa tra -1 e 1). Per evitare che l'algoritmo sia "influenzato" dalle feature con valori numerici più grandi, applicheremo una normalizzazione. Questo processo "schiaccia" tutte le feature in un intervallo comune, garantendo che ognuna di esse abbia la stessa opportunità di influenzare il risultato finale. Per completezza, utilizzerò 2 scaler: 

 - `MinMaxScaler` Questo scaler schiaccia tutti i dati in un intervallo da 0 a 1. Il valore più basso è assegnato a 0 mentre quello più alto invece è a 1. Suppone che ci sia una relazione lineare tra i dati
 - `StandardScaler` Questo scaler suppone una distribuzione gaussiana tra i dati e li porta ad un intervallo comune. La nuova scala di dati raggiunti avrà come media 0 e come deviazione standard 1

Inoltre, cosiderando che abbiamo tanti inquinanti, scelgo 3 inquinanti come core set. Quelli scelti sono: PM2.5, PM10 e NO2. I primi 2 indicano la presenza di microparticelle pericolose nell'aria, mentre l'ultima è un dato utile per l'analisi dell'inquinamento

Inanzitutto adesso rimuovo tutte le righe nella tabella gold relative a stazioni che non misurano tutti e 3 i parametri 

```python
print("\nInizio Fase 2: Preparazione dei dati per il ML...")
# 1. elimino la colonna __index_level_0__ che non serve
if '__index_level_0__' in df_profiles.columns:
    df_profiles.drop(columns=['__index_level_0__'], inplace=True)
    print("Colonna '__index_level_0__' rimossa.")



# 2. Definisco gli inquinanti "core" che ogni stazione DEVE misurare
core_pollutants_names = ['PM2.5', 'PM10', 'NO2_Biossido_di_azoto'] 
core_mean_cols = [f'mean_{name}' for name in core_pollutants_names if f'mean_{name}' in df_profiles.columns]
print(f"\nInquinanti 'core' scelti: {core_pollutants_names}")

# 3. Filtra df_profiles per tenere solo le stazioni che misurano TUTTI gli inquinanti core
mask_core_stations = (df_profiles[core_mean_cols] > 0).all(axis=1)
df_filtered_profiles = df_profiles[mask_core_stations].copy() # DataFrame intermedio con solo stazioni valide
display(df_filtered_profiles)
print(f"\nNumero stazioni totali: {len(df_profiles)}")
print(f"Numero stazioni nel 'Core Set': {len(df_filtered_profiles)}")
```

# Pulizia delle colonne
Dopo aver rimosso le righe che non ci interessano divido il dataframe in 2 parti, una con metadati e una con soli valori. 

Tratto `giorni_superamento_soglia_PM10` e `stazioni_vicine_20km` come metadati perchè li userò comunque per una validazione a posteriori

```python
print("\n[Passo 2.2] Separazione in DataFrame 'df_metadata' e 'df_core'...")

# 1. Definiamo le colonne di metadati
metadata_cols = [
    'giorni_superamento_soglia_PM10', 'stazioni_vicine_20km', 'Stazione', 'COMUNE', 
    'PROVINCIA', 'Altezza','Coord_X', 'Coord_Y', 'LON_GEO', 'LAT_GEO', 'Tipo_Area'
]
# Facciamo un controllo sui metadati scelti
metadata_cols_existing = [col for col in metadata_cols if col in df_filtered_profiles.columns]

# 2. Crea il DataFrame dei metadati
df_metadata = df_filtered_profiles[metadata_cols_existing].copy()

# 3. Identifica e seleziona le colonne relative ai "core pollutants"
# Questa è la logica chiave: una colonna è rilevante se il suo nome TERMINA con il nome di un inquinante scelto. 
# giorni superamento soglia PM10 lo considero come metadato perchè rompe la regola
core_data_cols = [
    col for col in df_filtered_profiles.columns 
    if any(col.endswith(pollutant) and col != 'giorni_superamento_soglia_PM10' for pollutant in core_pollutants_names)
]


# 4. Crea il DataFrame 'df_core' con solo i dati degli inquinanti scelti
df_core = df_filtered_profiles[core_data_cols].copy()

# VERIFICA FINALE 

print("\n--- Risultati della Trasformazione ---")
print("\ndf_metadata: Contiene i metadati per le stazioni del 'Core Set'.")
# display(df_metadata.head())
print(df_metadata.info())


print("\ndf_core: Contiene SOLO i dati relativi agli inquinanti 'core' per le stazioni filtrate.")
print(df_core.info())
display(df_core)

write_deltalake(
    DELTA_GOLD_TABLE,
    df_core,
    mode='overwrite' 
)

print(f"SCRITTURA COMPLETATA. La tabella {DELTA_GOLD_TABLE} è stata creata/aggiornata.")

```

## Fase 3: Primo Tentativo di Clustering

Ok, qui parte il primo vero tentativo di clustering. L'idea: 
1. Prendo le 21 feature "comportamentali" che ho costruito.
2. Le scalo (provo sia `StandardScaler` che `MinMaxScaler` per vedere se ci sono sostanziali differenze).
3. Eseguo K-Means per un range di K (da 2 a 10) e guardo i grafici di Inerzia e Silhouette per trovare il K migliore.


```python
print("[Passo 3] Preparazione Dati Comportamentali per il Clustering...")

# df_core contiene le 21 feature che useremo

scaler_std = StandardScaler()
scaler_mm = MinMaxScaler()
X_scaled_core_std = scaler_std.fit_transform(df_core)
X_scaled_core_mm = scaler_mm.fit_transform(df_core)

print(f"Matrice di feature standardizzata. Dimensioni std: {X_scaled_core_std.shape}     Dimensioni minmax:{X_scaled_core_mm.shape}")
#print(X_scaled_core.info())
#display(X_scaled_core)
print("I dati sono pronti per la ricerca del K ottimale.")
print("\nPasso 3 completato.")
```

---
## Ricerca del Numero Ottimale di Cluster (K)

La sfida principale dell'algoritmo K-Means è che dobbiamo specificare a priori il numero di cluster `K` da cercare. Scegliere un valore di `K` a caso non è un approccio scientifico.

Per prendere una decisione informata, eseguiremo un esperimento sistematico: testeremo una serie di valori di `K` (ad esempio, da 2 a 10) e valuteremo la qualità dei risultati usando due metriche standard:

*   **Metodo del Gomito (Elbow Method):** Misura la compattezza totale dei cluster attraverso un valore chiamato "Inertia". Cercheremo il "gomito", ovvero il punto nel grafico in cui aggiungere un nuovo cluster non porta più a un miglioramento significativo.

*   **Punteggio di Silhouette (Silhouette Score):** Misura la qualità complessiva dei cluster, valutando quanto ogni punto sia simile al proprio cluster rispetto agli altri. Un punteggio più vicino a 1 indica cluster densi e ben separati. Cercheremo il valore di `K` che massimizza questo punteggio.

Il codice seguente eseguirà questo ciclo di test e memorizzerà i risultati per la visualizzazione.

```python
k_range = range(2, 15)
inertias_std = []
silhouette_scores_std = []

inertias_mm = []
silhouette_scores_mm = []

print("\n-------Test su standard scaler-------")

for k in k_range:
    # Usiamo n_init='auto' o 10 per evitare warning futuri
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled_core_std) # ADDESTRIAMO SUI DATI COMPLETI
    inertias_std.append(kmeans.inertia_)
    silhouette_scores_std.append(silhouette_score(X_scaled_core_std, kmeans.labels_))
    # Stampiamo un feedback per ogni passo del ciclo, così vediamo il progresso.
    print(f"Testato K={k}: Inertia = {inertias_std[-1]:.2f}, Silhouette Score = {silhouette_scores_std[-1]:.4f}")

print("\n-------Test su minmax scaler-------")

for k in k_range:
    # Usiamo n_init='auto' o 10 per evitare warning futuri
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled_core_mm) # ADDESTRIAMO SUI DATI COMPLETI
    inertias_mm.append(kmeans.inertia_)
    silhouette_scores_mm.append(silhouette_score(X_scaled_core_mm, kmeans.labels_))
    # Stampiamo un feedback per ogni passo del ciclo, così vediamo il progresso.
    print(f"Testato K={k}: Inertia = {inertias_mm[-1]:.2f}, Silhouette Score = {silhouette_scores_mm[-1]:.4f}")

print("\nRicerca completata")
```

## La Scoperta Inaspettata: un Segnale Fortissimo per K=2

L'analisi della Silhouette ora mostra un segnale **inequivocabile**:
- **Per K=2**, ottengo uno score altissimo (quasi **0.63**), che indica una divisione quasi perfetta in due gruppi.
- **Per K > 2**, lo score crolla a metà, segno che forzare più cluster rompe la struttura naturale dei dati.

**Domanda:** Cosa può dividere in modo così netto le mie 26 stazioni? La mia prima ipotesi è la geografia: **pianura vs. montagna**.

```python
print("\n[Passo 4] Creazione della visualizzazione comparativa")

def plot_scaler_comparison(k_range, inertias_std, silhouette_scores_std, inertias_mm, silhouette_scores_mm, title="Confronto tra StandardScaler e MinMaxScaler"):
    """
    Genera una griglia di 4 grafici per confrontare i risultati del clustering 
    ottenuti con StandardScaler e MinMaxScaler.

    Args:
        k_range (range): L'intervallo di valori K testati.
        inertias_std (list): Lista delle inerzie ottenute con StandardScaler.
        silhouette_scores_std (list): Lista dei punteggi di silhouette con StandardScaler.
        inertias_mm (list): Lista delle inerzie ottenute con MinMaxScaler.
        silhouette_scores_mm (list): Lista dei punteggi di silhouette con MinMaxScaler.
        title (str, optional): Il titolo principale per la figura. 
                               Default "Confronto tra StandardScaler e MinMaxScaler".
    """
    
    # Creiamo una figura e una griglia di subplot 2x2.
    fig, ax = plt.subplots(2, 2, figsize=(20, 14))
    fig.suptitle(title, fontsize=22)

    # --- Riga Superiore: StandardScaler ---

    # Grafico 1 (in alto a sinistra): Gomito con StandardScaler
    ax[0, 0].plot(k_range, inertias_std, 'bo-', linewidth=2, markersize=8)
    ax[0, 0].set_title('StandardScaler - Metodo del Gomito', fontsize=16)
    ax[0, 0].set_xlabel('Numero di Cluster (K)', fontsize=12)
    ax[0, 0].set_ylabel('Inerzia', fontsize=12)
    ax[0, 0].grid(True, linestyle='--', alpha=0.6)

    # Grafico 2 (in alto a destra): Silhouette con StandardScaler
    ax[0, 1].plot(k_range, silhouette_scores_std, 'ro-', linewidth=2, markersize=8)
    ax[0, 1].set_title('StandardScaler - Analisi della Silhouette', fontsize=16)
    ax[0, 1].set_xlabel('Numero di Cluster (K)', fontsize=12)
    ax[0, 1].set_ylabel('Silhouette Score', fontsize=12)
    ax[0, 1].grid(True, linestyle='--', alpha=0.6)

    # --- Riga Inferiore: MinMaxScaler ---

    # Grafico 3 (in basso a sinistra): Gomito con MinMaxScaler
    ax[1, 0].plot(k_range, inertias_mm, 'go-', linewidth=2, markersize=8)
    ax[1, 0].set_title('MinMaxScaler - Metodo del Gomito', fontsize=16)
    ax[1, 0].set_xlabel('Numero di Cluster (K)', fontsize=12)
    ax[1, 0].set_ylabel('Inerzia', fontsize=12)
    ax[1, 0].grid(True, linestyle='--', alpha=0.6)

    # Grafico 4 (in basso a destra): Silhouette con MinMaxScaler
    ax[1, 1].plot(k_range, silhouette_scores_mm, 'mo-', linewidth=2, markersize=8)
    ax[1, 1].set_title('MinMaxScaler - Analisi della Silhouette', fontsize=16)
    ax[1, 1].set_xlabel('Numero di Cluster (K)', fontsize=12)
    ax[1, 1].set_ylabel('Silhouette Score', fontsize=12)
    ax[1, 1].grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

plot_scaler_comparison(k_range,inertias_std,silhouette_scores_std,inertias_mm,silhouette_scores_mm)
print("\nVisualizzazione completata.")
```

## Fase di Indagine: Verifichiamo l'Ipotesi con i Metadati

Se la mia ipotesi è corretta, dovrei vedere una differenza netta nell'altezza delle stazioni tra i due cluster.

Per fortuna, ho tenuto da parte la colonna `'Altezza'` nel `df_metadata`. Il piano è:
1. Prendere i risultati del clustering (le etichette 0 e 1 per ogni stazione).
2. Unirli ai metadati.
3. **Confrontare la distribuzione dell'altezza** per il Cluster 0 e il Cluster 1.

Mi aspetto di vedere un gruppo a bassa quota e uno ad alta quota.

```python
# FASE 5: CLUSTERING FINALE E VERIFICA DELL'IPOTESI SULL'ALTEZZA

print("\n[Passo 5] Esecuzione del Clustering Finale e Verifica dell'Ipotesi...")

# Dalle analisi precedenti, K=2 è risultato il numero di cluster ottimale.
# Scegliamo StandardScaler perché ha restituito il Silhouette Score marginalmente più alto.
k_ottimale = 2
X_scaled_final = X_scaled_core_std

print(f"Scelta finale: K={k_ottimale} con StandardScaler.")

# 1. Addestramento del modello K-Means definitivo
kmeans_final = KMeans(n_clusters=k_ottimale, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_scaled_final)

# 2. Creazione di un DataFrame di analisi per l'interpretazione
# Uniamo i metadati (che contengono 'Altezza') con i risultati del clustering
df_analysis = df_metadata.copy()
df_analysis['cluster'] = cluster_labels

# FASE 6: ANALISI DELL'ALTEZZA DEI CLUSTER
print("\n[Passo 6] Analisi dell'Altezza dei Cluster Trovati...")

# --- Analisi Numerica ---
print("\n--- Statistiche Descrittive dell'Altezza per Cluster ---")
altezza_stats = df_analysis.groupby('cluster')['Altezza'].describe()
print(altezza_stats)
print("-------------------------------------------------------")

# --- Calcolo e Stampa dell'Altezza Media ---
print("\n--- Altezza Media per Ciascun Cluster ---")
altezza_media = df_analysis.groupby('cluster')['Altezza'].mean()
for cluster_id, media in altezza_media.items():
    print(f"Altezza media del Cluster {cluster_id}: {media:.2f} metri s.l.m.")
print("-----------------------------------------")

# --- Analisi Visuale (con correzione per il FutureWarning) ---
print("\n--- Creazione Grafico Comparativo delle Altezze ---")
plt.figure(figsize=(12, 8))

sns.boxplot(x='cluster', y='Altezza', data=df_analysis, palette='viridis', width=0.4, 
            hue='cluster', legend=False) 

sns.swarmplot(x='cluster', y='Altezza', data=df_analysis, color='black', alpha=0.7, size=8)

plt.title('Confronto dell\'Altezza tra i 2 Cluster Identificati', fontsize=16)
plt.xlabel('Cluster ID', fontsize=12)
plt.ylabel('Altezza (metri s.l.m.)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## La Diagnosi: Trovato il Colpevole! È un Outlier.

L'analisi dell'altezza ha confermato l'ipotesi, ma in un modo ancora più estremo di quanto pensassi. I risultati sono chiarissimi:

- **Cluster 0:** Contiene **25 stazioni**, con un'altezza media di soli **46 metri**.
- **Cluster 1:** Contiene **1 sola stazione**, quella a **970 metri**.

**Conclusione:** Il clustering non ha trovato due "gruppi", ha semplicemente **isolato l'outlier**. La stazione montana è così diversa da tutte le altre che l'algoritmo ha trovato matematicamente più efficiente dedicarle un intero cluster.

Questo risultato, sebbene corretto, non è utile perché maschera tutti i pattern più sottili presenti nelle altre 25 stazioni.
Noto inoltre che ci sono altre 3 stazioni con altitudine sopra i 100 metri. Provo ad escludere anch'esse nel clustering per vedere se trovo miglioramenti

## Nuovo Piano d'Azione: Isolare l'Outlier e Rieseguire l'Analisi

Ok, ora che so qual è il problema, la soluzione è semplice e metodologicamente corretta.

**Nuovo Obiettivo:** Trovare i cluster significativi all'interno del gruppo omogeneo di stazioni di pianura/bassa collina.

Il piano è:
1. **Isolare l'Outlier:** Metto da parte la stazione a 970m. Sarà la prima categoria della mia tassonomia finale: la "Stazione di Fondo Montano".
2. **Creare un Nuovo Dataset:** Creo un `df_core_filtered` con le rimanenti 25 stazioni.
3. **Rieseguire il Clustering:** Rieseguo da capo la ricerca del K ottimale su questo nuovo dataset pulito.

Ora mi aspetto di trovare dei pattern molto più interessanti e meno banali.

### **Passo 1: Isolamento dell'Outlier e Preparazione del Nuovo Dataset**

**Obiettivo:** Rimuovere la stazione montana e preparare un nuovo set di dati pulito e pronto per l'analisi.

**Logica:** Identifichiamo la stazione con l'altitudine massima e la escludiamo da `df_core` e `df_metadata`. Successivamente, rieseguiamo lo scaling solo su questo nuovo gruppo di 25 stazioni, poiché le loro caratteristiche statistiche (media, deviazione standard) sono diverse da quelle del gruppo originale.







### **Passo 2: Nuova Ricerca del K Ottimale (sul Dataset Filtrato)**

**Obiettivo:** Eseguire l'analisi del gomito e della silhouette sul nuovo gruppo omogeneo di 25 stazioni per trovare il numero di cluster significativo al loro interno.

**Logica:** Liberi dall'influenza schiacciante dell'outlier, ora possiamo aspettarci di vedere emergere i pattern più sottili. I grafici di K-Means dovrebbero rivelare una struttura molto più interessante.


```python

print("\nIsolamento della Stazione Outlier...")

num_stazioni_eliminare=2

outlier_stations = df_metadata['Altezza'].nlargest(num_stazioni_eliminare)
outlier_ids = outlier_stations.index
outlier_info = df_metadata.loc[outlier_ids, ['Stazione', 'Altezza']]

print("stazioni outlier:")
for st_id, row in outlier_info.iterrows():
    print(f"- '{row['Stazione']}' (ID: {st_id}) a {row['Altezza']}m.")

# 2. Creiamo i nuovi DataFrame "_filtered", escludendo tutte le outlier
df_core_filtered = df_core.drop(index=outlier_ids)
df_metadata_filtered = df_metadata.drop(index=outlier_ids)

print(f"Nuovo set di dati creato con {len(df_core_filtered)} stazioni omogenee.")
print(f"Le {num_stazioni_eliminare} stazioni outlier sono state messe da parte per essere trattate come una categoria a sé.")



print("\nRiesecuzione dell'esperimento di scaling sul nuovo dataset")


X_scaled_std_filtered = scaler_std.fit_transform(df_core_filtered)


X_scaled_mm_filtered = scaler_mm.fit_transform(df_core_filtered)

k_range = range(2, 15) 

# Liste per i risultati
inertias_std_filtered = []
silhouette_scores_std_filtered = []
inertias_mm_filtered = []
silhouette_scores_mm_filtered = []

# --- Test su StandardScaler ---
print("\n--- Analisi con StandardScaler (su dati filtrati) ---")
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled_std_filtered)
    inertias_std_filtered.append(kmeans.inertia_)
    silhouette_scores_std_filtered.append(silhouette_score(X_scaled_std_filtered, kmeans.labels_))
    print(f"Testato K={k}: Inertia = {inertias_std_filtered[-1]:.2f}, Silhouette Score = {silhouette_scores_std_filtered[-1]:.4f}")

# --- Test su MinMaxScaler ---
print("\n--- Analisi con MinMaxScaler (su dati filtrati) ---")
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled_mm_filtered)
    inertias_mm_filtered.append(kmeans.inertia_)
    silhouette_scores_mm_filtered.append(silhouette_score(X_scaled_mm_filtered, kmeans.labels_))
    print(f"Testato K={k}: Inertia = {inertias_mm_filtered[-1]:.2f}, Silhouette Score = {silhouette_scores_mm_filtered[-1]:.4f}")

print("\nAnalisi completata. Ora visualizziamo i nuovi risultati.")

print("\n[Visualizzazione Pre-PCA] Confronto dei risultati degli scaler in 4 riquadri...")
plot_scaler_comparison(k_range,inertias_std_filtered,silhouette_scores_std_filtered,inertias_mm_filtered,silhouette_scores_mm_filtered)


```

### Il Percorso Investigativo e la Necessità di un Approccio Multivariato (PCA)

L'obiettivo di questa fase è trovare una suddivisione significativa delle stazioni in cluster omogenei. Un primo tentativo, che includeva tutte le 26 stazioni del "core set", ha prodotto un risultato statisticamente forte ma metodologicamente debole: una divisione in **K=2** con un Silhouette Score molto alto (~0.63). Un'analisi più approfondita ha rivelato che questo risultato era dominato da un singolo outlier, una stazione montana a 970m, che veniva isolata in un cluster a sé.

Per superare questo problema, è stato adottato un approccio iterativo di esclusione degli outlier basato sull'altitudine, con i seguenti risultati chiave:

1.  **Rimuovendo 1 Outlier (970m):** L'analisi è diventata ambigua. Il metodo del Gomito suggeriva un K=4, mentre l'analisi della Silhouette mostrava un picco a K=3 (con un valore modesto di ~0.29). Questa discrepanza tra le metriche è un forte indicatore che i cluster non sono ancora ben definiti.

2.  **Rimuovendo 2 Outlier (>200m):** Il segnale è cambiato di nuovo. Il Silhouette Score per K=2 è aumentato notevolmente (fino a ~0.42), indicando che anche la seconda stazione più alta aveva un'influenza sproporzionata. Tuttavia, i punteggi per K>2 sono rimasti bassi, suggerendo che i sottogruppi più fini erano ancora mascherati.

3.  **Rimuovendo Ulteriori Outlier (>100m):** Non sono stati osservati miglioramenti significativi, indicando che abbiamo raggiunto il limite dell'approccio basato sulla sola altitudine.

**Conclusione dell'Analisi Esplorativa:**
L'analisi manuale degli outlier ha avuto successo nel pulire il segnale più grossolano, ma ha anche dimostrato i suoi limiti. I "veri" outlier potrebbero non essere estremi su una singola variabile come l'altezza, ma sulla loro **combinazione unica di 21 feature comportamentali**. Continuare a rimuovere manualmente i punti diventa soggettivo e rischioso.

Pertanto, per superare questo limite e analizzare la struttura *multivariata* dei dati, si passa alla **Principal Component Analysis (PCA)**. L'obiettivo della PCA è ridurre il rumore e la complessità delle 21 dimensioni, estraendo le "super-feature" (componenti principali) che catturano l'essenza della varianza nei dati. Questo ci permetterà di eseguire il clustering su un segnale più pulito e potente, aumentando la probabilità di scoprire cluster significativi e interpretabili che prima erano nascosti.

Il dataset che userò per la pca è quello che esclude le 2 stazioni con l'altitudine maggiore


```python
from sklearn.decomposition import PCA

print("\n[Strategia Finale - Passo 1] Definizione del set di dati omogeneo di pianura...")


# 1. Identifichiamo le 2 stazioni con l'altitudine maggiore da escludere
outliers_ids_final = df_metadata.nlargest(2, 'Altezza').index

print("Le seguenti 2 stazioni saranno trattate come categorie separate ed escluse dall'analisi di clustering finale:")
print(df_metadata.loc[outliers_ids_final][['Stazione', 'Altezza']])

# 2. Rinomino i dataframe per migliore comprensione
df_core_pianura = df_core_filtered
df_metadata_pianura = df_metadata_filtered

print(f"\nSet di dati finale per il clustering creato con {len(df_core_pianura)} stazioni di pianura.")

# ===================================================================
# FASE FINALE - PASSO 2: APPLICAZIONE DELLA PCA
# ===================================================================
print("\n[Strategia Finale - Passo 2] Applicazione della PCA...")

# 1. Standardizziamo i dati (usando StandardScaler come best practice)
X_scaled_pianura = scaler_std.fit_transform(df_core_pianura)

# 2. Applichiamo la PCA
pca = PCA(n_components=0.90, random_state=42) # Aumentiamo al 90% per non perdere segnale
X_pca_pianura = pca.fit_transform(X_scaled_pianura)

print(f"\nPCA applicata. Sono state scelte {pca.n_components_} componenti.")
print(f"Queste componenti spiegano il {pca.explained_variance_ratio_.sum():.2%} della varianza totale.")

# ===================================================================
# FASE FINALE - PASSO 3: RICERCA DI K SUI DATI PCA
# ===================================================================
print("\n[Strategia Finale - Passo 3] Ricerca del K Ottimale sui dati PCA...")
k_range = range(2, 15)
inertias_pca = []
silhouette_scores_pca = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca_pianura)
    inertias_pca.append(kmeans.inertia_)
    silhouette_scores_pca.append(silhouette_score(X_pca_pianura, kmeans.labels_))
    print(f"Testato K={k}: Silhouette Score (PCA) = {silhouette_scores_pca[-1]:.4f}")

# Visualizziamo i risultati finali
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))
fig.suptitle('Analisi di K Finale (sui Dati di Pianura dopo PCA)', fontsize=16)
ax1.plot(k_range, inertias_pca, 'bo-'); ax1.set_title('Metodo del Gomito (Post-PCA)'); ax1.grid(True)
ax2.plot(k_range, silhouette_scores_pca, 'ro-'); ax2.set_title('Analisi della Silhouette (Post-PCA)'); ax2.grid(True)
plt.show()
```

## Confronto tra scalers
Per essere sicuri che i 2 scaler siano effettivamente uguali faccio un piccolo test modulando i componenti della pca

```python
print("\n[Esperimento 1] Confronto Scaler con n_components=4...")

# Dati di partenza
k_range = range(2, 15)
n_components_test = 4

# --- Preparazione Dati ---
# Scaler 1
scaler_std = StandardScaler()
X_scaled_std = scaler_std.fit_transform(df_core_pianura)
pca_std = PCA(n_components=n_components_test, random_state=42)
X_pca_std = pca_std.fit_transform(X_scaled_std)

# Scaler 2
scaler_mm = MinMaxScaler()
X_scaled_mm = scaler_mm.fit_transform(df_core_pianura)
pca_mm = PCA(n_components=n_components_test, random_state=42)
X_pca_mm = pca_mm.fit_transform(X_scaled_mm)

# --- Calcolo K-Means ---
inertias_std, silhouette_scores_std = [], []
inertias_mm, silhouette_scores_mm = [], []

for k in k_range:
    # StandardScaler
    kmeans_std = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X_pca_std)
    inertias_std.append(kmeans_std.inertia_)
    silhouette_scores_std.append(silhouette_score(X_pca_std, kmeans_std.labels_))
    # MinMaxScaler
    kmeans_mm = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X_pca_mm)
    inertias_mm.append(kmeans_mm.inertia_)
    silhouette_scores_mm.append(silhouette_score(X_pca_mm, kmeans_mm.labels_))

# --- Visualizzazione ---
plot_scaler_comparison(k_range, inertias_std, silhouette_scores_std, inertias_mm, silhouette_scores_mm,
                       title=f"Confronto Scaler con n_components={n_components_test}")
```

## Differenze tra i 2 scaler

Noto che ci sono grosse differenze tra i punteggi silhouette tra i 2 scalers. Considero lo standard scaler perchè migliore

```python
# ===================================================================
# ESPERIMENTO 2: ANALISI DI SENSIBILITÀ SU N_COMPONENTS
# ===================================================================
print("\n[Esperimento 2] Analisi di Sensibilità su n_components...")

# Dati di partenza (scalati con StandardScaler)
X_scaled = StandardScaler().fit_transform(df_core_pianura)
k_range = range(2, 15)

# Testiamo con 2, 3 e 4 componenti
for n_comp in [2, 3, 4]:
    print(f"\n--- Analisi in corso con n_components={n_comp} ---")
    
    # Applichiamo PCA con il numero corrente di componenti
    pca = PCA(n_components=n_comp, random_state=42)
    X_pca = pca.fit_transform(X_scaled)
    
    print(f"Varianza spiegata con {n_comp} componenti: {pca.explained_variance_ratio_.sum():.2%}")
    
    inertias = []
    silhouette_scores = []

    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X_pca)
        inertias.append(kmeans.inertia_)
        silhouette_scores.append(silhouette_score(X_pca, kmeans.labels_))

    # --- Plot dei risultati per questo n_comp ---
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))
    fig.suptitle(f'Analisi di K con n_components={n_comp}', fontsize=16)
    ax1.plot(k_range, inertias, 'bo-'); ax1.set_title('Metodo del Gomito'); ax1.grid(True)
    ax2.plot(k_range, silhouette_scores, 'ro-'); ax2.set_title('Analisi della Silhouette'); ax2.grid(True)
    plt.show()
```

# Considerazioni prima del clustering finale
Dopo aver applicato PCA I risultati sono molto più chiari. Il punto di gomito si vede a k=4 ed è quello con il punteggio di silhouette migliore.

1. **Clustering:** Eseguiremo K-Means sui dati a 4 dimensioni trasformati dalla PCA.

2. **Analisi per Cluster:**  
   Per ciascuno dei 4 cluster trovati, stamperemo due tabelle distinte:
   - **Composizione del Cluster:**  
     Una lista delle stazioni che ne fanno parte, arricchita da metadati chiave come il Comune e l'Altezza, per capire chi e dove.
   - **Profilo Comportamentale Medio:**  
     Le medie delle sole feature comportamentali, per capire la *"personalità"* del cluster.

3. **Visualizzazione Mappata:**  
   Creeremo un grafico di dispersione basato sulle prime due Componenti Principali per visualizzare la posizione relativa dei cluster, etichettando ogni punto per una facile identificazione.

   
Questo ci fornirà la tassonomia comportamentale finale per le stazioni di pianura della regione che abbiamo selezionato

```python
# X_pca_pianura è il df a cui ho applicato PCA
# k=4 numero di cluster da sceglier

# Parametri finali scelti dall'analisi precedente
k_ottimale = 4
X_final_for_clustering = X_pca_pianura # Dati a 4 dimensioni

print(f"[Fase Finale] Esecuzione del clustering con K={k_ottimale}...")

# 1. Addestramento del modello K-Means definitivo
kmeans_final = KMeans(n_clusters=k_ottimale, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_final_for_clustering)

# 2. Creiamo un DataFrame con i risultati per una facile analisi
df_risultati = df_metadata_pianura.copy()
df_risultati = df_risultati.join(df_core_pianura)
df_risultati['cluster'] = cluster_labels

# ===================================================================
# ANALISI DETTAGLIATA PER CIASCUN CLUSTER
# ===================================================================

for i in range(k_ottimale):
    print(f"\n\n{'='*25} ANALISI DEL CLUSTER {i} {'='*25}")
    
    # Selezioniamo solo le stazioni appartenenti a questo cluster
    cluster_df = df_risultati[df_risultati['cluster'] == i]
    
    # Tabella 1: COMPOSIZIONE DEL CLUSTER (Chi e Dove)
    print(f"\n[Cluster {i}] Composizione: {len(cluster_df)} stazioni")
    print("--------------------------------------------------")
    # Mostriamo metadati utili per l'interpretazione
    display(cluster_df[['Stazione', 'COMUNE', 'Altezza', 'giorni_superamento_soglia_PM10', 'stazioni_vicine_20km']])
    
    # Tabella 2: PROFILO COMPORTAMENTALE MEDIO (La "Personalità")
    print(f"\n[Cluster {i}] Profilo Comportamentale Medio:")
    print("--------------------------------------------------")
    # Calcoliamo la media solo delle feature comportamentali
    profilo_medio = cluster_df[df_core_pianura.columns].mean()
    display(profilo_medio.to_frame(name='Valore Medio').T)

# ===================================================================
# VISUALIZZAZIONE FINALE SULLA MAPPA 2D
# ===================================================================
print(f"\n\n{'='*25} VISUALIZZAZIONE FINALE DEI CLUSTER {'='*25}")
print("Il grafico seguente è una 'mappa' semplificata in 2D.")
print("Mostra la posizione relativa delle stazioni lungo le due direzioni di massima varianza (PC1 e PC2).")
print("I colori indicano il cluster assegnato dall'algoritmo che ha lavorato nello spazio a 4 dimensioni.")

# Usiamo le prime due componenti della PCA (già calcolate in X_pca_pianura)
df_visualization = pd.DataFrame({
    'PC1': X_pca_pianura[:, 0],
    'PC2': X_pca_pianura[:, 1],
    'cluster': cluster_labels,
    'Stazione': df_core_pianura.index
})

plt.figure(figsize=(16, 10))
sns.scatterplot(
    data=df_visualization, x='PC1', y='PC2', hue='cluster',
    palette='viridis', s=200, alpha=0.9, edgecolor='k'
)

# Aggiungiamo le etichette delle stazioni
for j in range(df_visualization.shape[0]):
    plt.text(x=df_visualization.PC1[j] + 0.1, y=df_visualization.PC2[j] + 0.1,
             s=df_visualization.Stazione[j], fontdict=dict(color='black', size=9))

plt.title(f'Mappa 2D dei {k_ottimale} Cluster di Pianura (Proiezione PCA)', fontsize=18)
plt.xlabel('Componente Principale 1 (Direzione di Massima Varianza)', fontsize=12)
plt.ylabel('Componente Principale 2 (Seconda Direzione di Massima Varianza)', fontsize=12)
plt.grid(True); plt.legend(title='Cluster ID'); plt.show()
```
